```python
# Data input: data_London.csv(11 GB);
# Data output: data_London/*.csv, data_London_2013_hh/*.csv, data_London_attr.csv, data_London_attr_2013.csv;

# Load data (London Smart meter)
import pandas as pd

def path(id):
    return '../data/data_London/' + id + '.csv'

def path_2013_hh(id):
    return '../data/data_London_2013_hh/' + id + '.csv'

data_path = '../data/data_London.csv'
df = pd.read_csv(data_path)
```

```python
df.columns
```

```python
df.columns = ['ID', 'Tariff', 'DateTime', 'KW', 'Acorn', 'Acorn_grouped']
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Adjust the data type of some columns
df['DateTime'] = df['DateTime'].astype('datetime64[ns]')
df = df[df['KW']!='Null']
df['KW'] = df['KW'].astype('float')
```

```python
df.dtypes
```

```python
# Generate the attribute list
attr = df[['ID', 'Tariff', 'Acorn', 'Acorn_grouped']]
attr.drop_duplicates(inplace=True)
attr.sort_values(by='ID', inplace=True)
attr.reset_index(drop=True, inplace=True)
attr.to_csv('../data/data_London_attr.csv', index=None)
```

```python
# Generate the separate household load data
from tqdm import tqdm

for i in tqdm(range(len(attr))):
    tmp = df[df['ID'] == attr['ID'][i]]
    tmp.drop_duplicates(inplace=True)
    tmp = tmp[['DateTime', 'KW']]
    tmp.to_csv(path(attr['ID'][i]), index=None)
```

```python
# Generate the attribute list with Tariff = Std and enough data in 2013
attr = attr[(attr['Tariff'] == 'Std') & (attr['Acorn_grouped'] != 'ACORN-U')]
attr.reset_index(drop=True, inplace=True)

num_list = []

for i in range(len(attr)):
    df = pd.read_csv(path(attr['ID'][i]))
    if len(df[(df['DateTime'] > '2013-01-01 00:00:00') & (df['DateTime'] <= '2014-01-01 00:00:00')]) < 17515:
        num_list.append(attr['ID'][i])
```

```python
# Delete the useless ID
for i in range(len(num_list)):
    attr = attr[attr['ID'] != num_list[i]]

attr.reset_index(drop=True, inplace=True)
attr.to_csv('../data/data_London_attr_2013.csv', index=None)
```

```python
# Missing values
# Generate the separate load data (half hour)
import dateutil
import pandas as pd

for i in tqdm(range(len(attr))):
    df = pd.read_csv(path(attr['ID'][i]))
    # Select the load data in 2013
    df = df[(df['DateTime'] > '2013-01-01 00:00:00') & (df['DateTime'] <= '2014-01-01 00:00:00')]
    
    # Construct the empty DataFrame to store the interpolation
    add = pd.DataFrame(columns=['DateTime', 'KW'])
    
    for j in range(365*48):
        t = pd.to_datetime('2013-01-01 00:00:00') + dateutil.relativedelta.relativedelta(minutes=30*(j+1))
        if str(t) not in df['DateTime'].values:
            
            # Use average values as interpolation
            value = (df[df['DateTime'] > str(t)].iloc[0, 1] + df[df['DateTime'] < str(t)].iloc[-1, 1])/2
            tmp = pd.DataFrame([[str(t), value]], columns=['DateTime', 'KW'])
            add = pd.concat([add, tmp], axis=0)
    
    df = pd.concat([df, add], axis=0)
    df.sort_values(by='DateTime').to_csv(path_2013_hh(attr['ID'][i]), index=None)
```
